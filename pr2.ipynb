```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import SimpleImputer, KNNImputer , IterativeImputer
from scipy import stats
```

```python
df = pd.read_csv('patient_health_500.csv')
df
```

```python
df.isnull().sum()
```

```python
(df.isnull().sum()/len(df))*100
```

```python
simple_im  = SimpleImputer(strategy='mean')
df["bmi"] = simple_im.fit_transform(df[["bmi"]])
print(df)
```

```python
imputer_mf = SimpleImputer(strategy='most_frequent')
df["region"] = imputer_mf.fit_transform(df[["region"]]).ravel()
print(df)
```

```python
imputer_mf1 = SimpleImputer(strategy='most_frequent')
df['gender'] = imputer_mf1.fit_transform(df[["gender"]]).ravel()
print(df)
```

```python
columns = ["age", "bmi", "blood_pressure", "cholesterol", "glucose"]
```

```python
for col in columns:
    df[col + '_missing'] = df[col].isnull().astype(int)
    missing_val = df[col].isnull().sum()
    if missing_val > 0:
        random_sample = df[col].dropna().sample(missing_val, random_state=42) 
        random_sample.index = df[df[col].isnull()].index
        df.loc[df[col].isnull(), col] = random_sample
```

```python
knn = KNNImputer(n_neighbors=5)

df[columns] = knn.fit_transform(df[columns])
print(df)
```

```python
mice_imputer = IterativeImputer(random_state=0)

df[columns] = mice_imputer.fit_transform(df[columns])

df.isnull().sum()
```

```python
z_scores = np.abs(stats.zscore(df[["cholesterol", "glucose"]]))

outliers = (z_scores > 3).any(axis=1)
```

```python
print((outliers).sum())
```

```python
extreme_patients = df[(z_scores > 3).any(axis=1)]

extreme_patients
```

```python
Q1 = df["bmi"].quantile(0.25)
Q3 = df["bmi"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower: ", lower)
print("Upper: ", upper)
```

```python
outliers_IQR = df[(df["bmi"] < lower) | (df["bmi"] > upper)]

outliers_IQR
```

```python
for col in columns:
  lower= df[col].quantile(0.01)
  upper= df[col].quantile(0.99)
```

```python
df = (df[col] >= lower) & (df[col] <= upper)
```
